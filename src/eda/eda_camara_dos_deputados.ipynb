```
# -*- coding: utf-8 -*-
import pyspark.sql.functions as F
```

```
%sql 

CREATE TABLE IF NOT EXISTS tb_expenses
USING PARQUET
OPTIONS (path "/FileStore/tables/curated/expenses/*.parquet");

CREATE TABLE IF NOT EXISTS tb_politicians
USING PARQUET
OPTIONS (path "/FileStore/tables/curated/politicians/*.parquet");

CREATE TABLE IF NOT EXISTS tb_propositions
USING PARQUET
OPTIONS (path "/FileStore/tables/curated/propositions/*.parquet");

CREATE TABLE IF NOT EXISTS tb_propositions_authors
USING PARQUET
OPTIONS (path "/FileStore/tables/curated/propositions_authors/*.parquet");

CREATE TABLE IF NOT EXISTS tb_propositions_themes
USING PARQUET
OPTIONS (path "/FileStore/tables/curated/propositions_themes/*.parquet");
```

```
%sql 
select 
  NM_PARLAMENTAR
  , NU_ANO
  , round(sum(VL_DOCUMENTO),2) VL_GASTO_TOTAL
from 
  tb_politicians 
inner join
  tb_expenses
on
  tb_politicians.ID_DEPUTADO = tb_expenses.ID_DEPUTADO
--where
  --NM_PARLAMENTAR LIKE '%BOLSONARO%'
group by
  NM_PARLAMENTAR
  , NU_ANO
order by
  NM_PARLAMENTAR
  , NU_ANO
```

```
%sql 
select 
  DC_TEMA
  , count(1) PROPOSICOES
from 
  tb_politicians 
inner join
  tb_propositions_authors
on
  tb_politicians.ID_DEPUTADO = tb_propositions_authors.ID_DEPUTADO
inner join
  tb_propositions
on
  tb_propositions.ID_PROPOSICAO = tb_propositions_authors.ID_PROPOSICAO
inner join
  tb_propositions_themes
on
  tb_propositions.ID_PROPOSICAO = tb_propositions_themes.ID_PROPOSICAO
group by
  DC_TEMA
order by
  DC_TEMA
```

```
%sql 
select 
  NM_PARLAMENTAR
  , tb_politicians.SG_PARTIDO
  , count(distinct tb_propositions.ID_PROPOSICAO) PROPOSICOES
  , round(sum(VL_DOCUMENTO),2) VL_GASTO_TOTAL
from 
  tb_politicians 
inner join
  tb_propositions_authors
on
  tb_politicians.ID_DEPUTADO = tb_propositions_authors.ID_DEPUTADO
inner join
  tb_propositions
on
  tb_propositions.ID_PROPOSICAO = tb_propositions_authors.ID_PROPOSICAO
inner join
  tb_propositions_themes
on
  tb_propositions.ID_PROPOSICAO = tb_propositions_themes.ID_PROPOSICAO
inner join
  tb_expenses
on
  tb_politicians.ID_DEPUTADO = tb_expenses.ID_DEPUTADO
where
  tb_expenses.NU_ANO = 2018
  and tb_propositions.NU_ANO = 2018
group by
  NM_PARLAMENTAR
  , tb_politicians.SG_PARTIDO
```
